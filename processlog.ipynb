```python
import pandas as pd
import numpy as np
```

```python
mylog = pd.read_excel(r'./HKU956/MyLog.xlsx')
av_rating = pd.read_csv(r'./HKU956/3. AV_ratings.csv')
```

```python
mylog.head()
```

```python
mylog['start_time'] = mylog['start_time'].values.astype(np.int64) // 10 ** 9
mylog['end_time'] = mylog['end_time'].values.astype(np.int64) // 10 ** 9
mylog['ts_d'] = mylog['end_time'] - mylog['start_time']
mylog.head()
```

```python
hr_262957 = pd.read_csv(r'./HKU956/1. physiological_signals/hku1903/HR/0_262957.csv', header=None)
len(hr_262957.iloc[:,0].values)
```

```python
mylog = mylog.groupby(['user_id']).apply(lambda x: x.sort_values('start_time', ascending=True))

song_no = []

for user in mylog['user_id'].unique():
    song_no.extend(range(mylog[mylog['user_id']==user].shape[0]))

mylog['song_no'] = song_no
```

```python
mylog.rename(columns={'user_id': 'participant_id'}, inplace=True)
mylog
```

```python
av_rating.merge(mylog, on=['participant_id', 'song_no', 'song_id'], how='left').isnull().values.any()
```
